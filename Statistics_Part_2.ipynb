1. What is hypothesis testing in statistics?

Hypothesis testing is a core concept in statistics used to make decisions or inferences about a population based on sample data. It provides a structured way to test assumptions (called hypotheses) using data and probability.



2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

- Null Hypothesis (H₀):

Definition:
A statement that there is no effect, no difference, or no relationship. It is the default or status quo assumption.

Purpose:
It is the hypothesis that the test attempts to disprove.

Example:
"The average test score of students is equal to 70."
⇒ H₀: μ = 70

- Alternative Hypothesis (H₁ or Ha):

Definition:
A statement that there is an effect, difference, or relationship. It reflects what the researcher aims to prove.

Purpose:
It is accepted if the null hypothesis is rejected.

Example:
"The average test score of students is not equal to 70."
⇒ H₁: μ ≠ 70



3. What is the significance level in hypothesis testing, and why is it important?
-The significance level in hypothesis testing, commonly denoted by α (alpha), is the threshold for determining whether a null hypothesis should be rejected. It represents the probability of making a Type I error, which is rejecting the null hypothesis when it is actually true.
Controls False Positives:

By setting a significance level, researchers limit the risk of wrongly claiming a statistically significant effect or difference when none exists.

1.Guides Decision-Making:

If the p-value (probability of observing the data if the null is true) is less than α, we reject the null hypothesis.

This provides a standardized rule for making decisions based on data.

2.Balances Risk:

3.Choosing α involves a trade-off:

Lower α reduces false positives but increases the risk of Type II error (failing to detect a real effect).

Researchers choose α based on the context and consequences of errors in the specific study.

4.Supports Reproducibility:

Using a predefined α ensures that results can be compared and replicated by other researchers under the same criteria.


4. What does a P-value represent in hypothesis testing?
- It measures how compatible your data is with the null hypothesis.

  A small p-value means the observed data is unlikely under H₀, suggesting evidence against H₀.

  A large p-value means the data is consistent with H₀, so there is no strong evidence against it.



5.  How do you interpret the P-value in hypothesis testing?
- The p-value helps you determine whether your test results are statistically significant and whether to reject the null hypothesis (H₀).

P-value ≤ α - Reject H₀ - significant result.

P-value > α - Fail to reject H₀ - not enough evidence.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- In hypothesis testing, a Type I error occurs when we reject the null hypothesis (H₀) even though it is actually true. This means we are detecting an effect or difference that doesn't really exist — a false positive. For example, concluding a drug is effective when it actually isn't.

A Type II error happens when we fail to reject the null hypothesis when it is actually false. This means we are missing a real effect — a false negative. For instance, concluding a drug has no effect when it actually does.

In short:

Type I error = false alarm.

Type II error = missed detection.

7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

-A one-tailed test is used when you want to determine if a parameter is either greater than or less than a certain value, focusing on one direction of change.

A two-tailed test is used when you want to determine if a parameter is simply different from a certain value, regardless of whether it is higher or lower, so it considers both directions.

In a one-tailed test, the entire significance level (α) is placed in one tail of the distribution (either left or right), while in a two-tailed test, the α is split equally between the two tails.

A one-tailed test has more power to detect an effect in one direction, but it cannot detect an effect in the opposite direction, whereas a two-tailed test can detect deviations in both directions but is slightly less powerful in either one.

You use a one-tailed test when your hypothesis is directional, such as testing whether a drug increases recovery speed. You use a two-tailed test when your hypothesis is non-directional, such as testing whether a drug has a different effect (could be better or worse).





8.What is the Z-test, and when is it used in hypothesis testing?

-A Z-test is a type of statistical test used to determine whether there is a significant difference between sample and population means (or between two sample means), assuming the data follows a normal distribution and the population standard deviation is known.

You use a Z-test when all of the following conditions are met:

The sample size is large (typically n ≥ 30).

The population is normally distributed (or the sample is large enough for the Central Limit Theorem to apply).

The population standard deviation (σ) is known.

11.What is the difference between a Z-test and a T-test?

-A Z-test is used when the population standard deviation (σ) is known and the sample size is large (typically n ≥ 30), whereas a T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30).

The Z-test relies on the standard normal distribution, while the T-test uses the t-distribution, which has fatter tails to account for greater uncertainty in small samples.

As the sample size increases, the t-distribution gradually approaches the normal distribution, meaning that with large samples, t-tests and z-tests give similar results.

In practice, the T-test is more commonly used because in real-world situations, the population standard deviation is usually not known.

The Z-test is more precise when the population standard deviation is known, but the T-test provides a better estimate when the standard deviation has to be estimated from the sample.



12.What is the T-test, and how is it used in hypothesis testing?

-A T-test is a statistical method used to determine whether there is a significant difference between two means — either between a sample and a population, or between two sample groups.
It is based on the t-distribution, which is used when the sample size is small (typically less than 30) and the population standard deviation is unknown.

1.State hypotheses:

   a. Null hypothesis (H₀): There is no difference between means.

b.Alternative hypothesis (H₁): There is a difference.

2.Choose significance level (commonly α = 0.05).

3.Calculate the t-statistic using sample data.

4.Find the critical t-value from the t-distribution table (based on degrees of freedom).

5.Compare the calculated t-value with the critical value:

a.If |t| > critical value → Reject H₀ (significant result)

b.If |t| ≤ critical value → Fail to reject H₀ (not significant)



13.What is the relationship between Z-test and T-test in hypothesis testing?

-Both the Z-test and T-test compare sample statistics to population parameters to test hypotheses about means.

-The main difference lies in whether the population standard deviation is known:

-If the population standard deviation is known, a Z-test is used.

-If the population standard deviation is unknown, a T-test is used, and the sample standard deviation is used as an estimate.

-The Z-test uses the standard normal distribution, while the T-test uses the t-distribution, which has fatter tails to account for the additional variability from estimating the standard deviation.

-The T-distribution becomes very similar to the normal distribution as the sample size increases, meaning the T-test approaches the Z-test for large samples (typically n ≥ 30).

-Both tests involve calculating a test statistic (Z or t), comparing it to a critical value, and making a decision about the null hypothesis.



14.What is a confidence interval, and how is it used to interpret statistical results?

-A confidence interval is a range of values that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence.

It reflects the degree of uncertainty or certainty in an estimate based on sample data.

How Is It Used to Interpret Statistical Results?

1.Estimates a Range for the Parameter:
A confidence interval provides a range, not just a single estimate, making it more informative than a point estimate.

2.Supports Decision-Making:
If a null value (like 0 or a known mean) falls outside the confidence interval, it suggests a statistically significant result.

3.Indicates Precision:

a.Narrower intervals mean the estimate is more precise.

b.Wider intervals suggest more uncertainty.

4.Helps Interpret Effect Size:

It gives insight into how large or small the true effect might be, which is useful beyond just determining significance.

15.What is the margin of error, and how does it affect the confidence interval?

The margin of error is the maximum expected difference between the true population parameter and a sample estimate.

It represents the range of uncertainty around a sample statistic due to sampling variability.

Mathematically, it’s the amount you add and subtract from the sample estimate to create a confidence interval.

How Does It Affect the Confidence Interval?

1.The margin of error determines the width of the confidence interval.

-Confidence Interval (CI) = Point estimate ± Margin of Error

2.A larger margin of error results in a wider confidence interval, indicating more uncertainty in the estimate.

3.A smaller margin of error results in a narrower confidence interval, showing more precision.

20.What is an ANOVA test, and what are its assumptions?

-ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences among the means of three or more independent groups.

Rather than comparing group means one by one (like with t-tests), ANOVA checks if at least one group mean is significantly different from the others.

Assumptions of ANOVA:

1.Independence of observations

Each sample or group must be collected independently.

2.Normality

The data in each group should be approximately normally distributed.

3.Homogeneity of variance (homoscedasticity)

a.The variances across groups should be roughly equal.

b.This can be tested using Levene’s test or Bartlett’s test.

4.Measurement scale

The dependent variable should be measured on an interval or ratio scale (i.e., numerical and continuous).

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    return p if x == 1 else 1 - p
print(bernoulli_pdf(1, 0.7))
print(bernoulli_pdf(0, 0.7))
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt
def calculate_and_plot_zscores(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(z_scores, marker='o', linestyle='-', color='purple')
    plt.title("Z-scores Line Plot")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.axhline(0, color='gray', linestyle='--')
    plt.grid(True)
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=20, edgecolor='black', color='skyblue', density=True)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    return z_scores
sample_data = np.random.normal(100, 15, 100)
z_scores = calculate_and_plot_zscores(sample_data)
```

```python
#15.Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm
def z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    sample = np.array(sample_data)
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    if alternative == 'two-sided':
        p = 2 * (1 - norm.cdf(abs(z)))
        crit_val = norm.ppf(1 - alpha / 2)
    elif alternative == 'greater':
        p = 1 - norm.cdf(z)
        crit_val = norm.ppf(1 - alpha)
    elif alternative == 'less':
        p = norm.cdf(z)
        crit_val = norm.ppf(alpha)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")
    reject = p < alpha
    result = "Reject H₀" if reject else "Fail to reject H₀"
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p:.4f}")
    print(f"Significance Level: {alpha}")
    print(f"Conclusion: {result}")
    return z, p, result
sample_data = [51, 49, 50, 52, 48, 53, 47, 50, 51, 49]
population_mean = 50
population_std = 2
alpha = 0.05
z_test(sample_data, population_mean, population_std, alpha, alternative='two-sided')
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats
def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin
    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Level: {confidence*100:.0f}%")
    print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {confidence*100:.0f}% confident the true population mean lies between {lower:.2f} and {upper:.2f}.")
    return (lower, upper)
sample_data = [12, 15, 14, 10, 13, 12, 14, 11, 13, 15]
confidence_interval(sample_data, confidence=0.95)
```

```python
#1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy.stats import norm
def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    sample_data = np.array(sample_data)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    standard_error = population_std / np.sqrt(n)
    z_score = (sample_mean - population_mean) / standard_error
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
    elif alternative == 'less':
        p_value = norm.cdf(z_score)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")
    if p_value < alpha:
        conclusion = "Reject the null hypothesis (H₀)."
    else:
        conclusion = "Fail to reject the null hypothesis (H₀)."
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Significance Level (α): {alpha}")
    print(f"Conclusion: {conclusion}")
    return z_score, p_value, conclusion
sample = [98, 102, 99, 101, 97, 100, 103, 98, 99, 101]
population_mean = 100
population_std = 5
alpha = 0.05
one_sample_z_test(sample, population_mean, population_std, alpha, alternative='two-sided')
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def two_tailed_z_test_with_plot(sample_data, population_mean, population_std, alpha=0.05):
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    n = len(sample_data)
    standard_error = population_std / np.sqrt(n)
    z_score = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    z_critical = norm.ppf(1 - alpha/2)
    if abs(z_score) > z_critical:
        conclusion = "Reject H₀"
    else:
        conclusion = "Fail to reject H₀"
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-value: ±{z_critical:.4f}")
    print(f"Conclusion: {conclusion}")
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color='blue')
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (left)')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region (right)')
    plt.axvline(z_score, color='green', linestyle='--', lw=2, label=f'Z = {z_score:.2f}')
    plt.axvline(-z_critical, color='black', linestyle='dotted', label=f'Z = ±{z_critical:.2f}')
    plt.axvline(z_critical, color='black', linestyle='dotted')
    plt.title("Two-Tailed Z-Test: Decision Regions")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
sample = [102, 98, 100, 97, 101, 99, 100, 103, 98, 100]
population_mean = 100
population_std = 5
alpha = 0.05
two_tailed_z_test_with_plot(sample, population_mean, population_std, alpha)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayesian_inference(prior, likelihood, false_positive, confidence=2):
    not_prior = 1 - prior
    evidence = (likelihood * prior) + (false_positive * not_prior)
    posterior = (likelihood * prior) / evidence
    print("=== Bayesian Inference Result ===")
    print(f"Prior P(Disease): {prior}")
    print(f"Likelihood P(Positive | Disease): {likelihood}")
    print(f"False Positive Rate P(Positive | No Disease): {false_positive}")
    print(f"Evidence P(Positive): {evidence:.4f}")
    print(f"Posterior P(Disease | Positive Test): {posterior:.4f}")
    return posterior
bayesian_inference(prior=0.01, likelihood=0.99, false_positive=0.05)
```

```python
#12.Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
def chi_square_test_independence():
    data = {'Like':   [20, 30],
            'Dislike':[15, 35]}
    table = pd.DataFrame(data, index=['Male', 'Female'])
    chi2, p, dof, expected = chi2_contingency(table)
    print("=== Contingency Table ===")
    print(table)
    print("\n=== Chi-Square Test Result ===")
    print(f"Chi-square statistic: {chi2:.4f}")
    print(f"Degrees of freedom: {dof}")
    print(f"P-value: {p:.4f}")
    print("\nExpected Frequencies:")
    print(pd.DataFrame(expected, index=table.index, columns=table.columns))
    if p < 0.05:
        print("\nConclusion: Reject H₀ → There is a significant association between the variables.")
    else:
        print("\nConclusion: Fail to reject H₀ → No significant association between the variables.")
chi_square_test_independence()
```

```python
#14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare
def goodness_of_fit_test(observed, expected=None):
    observed = np.array(observed)
    if expected is None:
        expected = np.full_like(observed, fill_value=observed.sum() / len(observed))
    chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)
    print("=== Chi-Square Goodness-of-Fit Test ===")
    print(f"Observed: {observed}")
    print(f"Expected: {expected}")
    print(f"Chi-square Statistic: {chi2_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Conclusion: Reject H₀ → The observed data does not fit the expected distribution.")
    else:
        print("Conclusion: Fail to reject H₀ → The observed data fits the expected distribution.")
    return chi2_stat, p_value
observed_rolls = [8, 9, 10, 11, 12, 10]
goodness_of_fit_test(observed_rolls)
```
